```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_filepath = "fifa.csv"

fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=fifa_data)
```

```python
list(fifa_data.columns)
```

```python
plt.title("Fifa Rankings")
sns.lineplot(data=fifa_data['GER'], label='Germany')
sns.lineplot(data=fifa_data['ARG'], label='Argentina')
```

```python
flight_delay_path = 'flight_delays.csv'
flight_delay_data = pd.read_csv(flight_delay_path, index_col="Month")
flight_delay_data.head()
```

```python
sns.set_theme(style='ticks', palette='pastel')
plt.figure(figsize=(10,6))
plt.title("Flight Delays")
sns.barplot(x=flight_delay_data.index, y=flight_delay_data['NK'])
plt.ylabel("Arrival Delay (in Minutes)")
```

```python
plt.figure(figsize=(14, 7))
plt.title("Avg Arrival Delay Each Month For Each Plane")

sns.heatmap(data=flight_delay_data, annot=True)
plt.xlabel("Airline")
```

```python
insurance_filepath = 'insurance.csv'
insurance_data = pd.read_csv(insurance_filepath)
insurance_data.head()
```

```python
x = insurance_data['bmi'] 
y = insurance_data['charges']
sns.scatterplot(x=x, y=y, hue=insurance_data['smoker'])
```

```python
sns.regplot(x=x, y=y)
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

```python
iris_data = pd.read_csv('iris.csv', index_col='Id')
iris_data.head()
```

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
sns.kdeplot(data=iris_data["Petal Length (cm)"], fill=True)
```

```python
sns.kdeplot(x='Petal Length (cm)', y='Sepal Width (cm)', data= iris_data, fill=True)
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'],kind='kde', fill =True)
```

```python
sns.histplot(x=iris_data['Petal Length (cm)'], hue=iris_data['Species'])
plt.title("Histogram of Petal Length in cm, by Species")
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Length (cm)'], hue=iris_data['Species'], fill=True)
```
